```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import folium
import re
```

```python
df = pd.read_csv('/home/g49/Documents/cholera/data/Cholera.csv')
df.head()
```

```python
def coords_grabber(str):
    coords = re.search('<Point><coordinates>(.+)</coordinates></Point>', str).group(1)
    coords = coords.split(',')
    lat = float(coords[1])
    long = float(coords[0])
    return (lat, long)
def lat(str):
    return coords_grabber(str)[0]
def lon(str):
    return coords_grabber(str)[1]
```

```python
lats = df['geometry'].apply(lat)
```

```python
df['lat'] = lats
```

```python
df['lon'] = df['geometry'].apply(lon)
```

```python
df = df.drop('geometry', axis=1)
```

```python
df_top = df[df['count'] >= 0]
df_bottom = df[df['count'] == -999]
```

```python
df_top.tail()
```

```python
df_bottom
```

```python
first_map = folium.Map(location = [51.513, -0.135], zoom_start=16.25, tiles='Stamen Watercolor')
```

```python
first_map
```

```python
for idx in df_top.index:
    lat = df_top.loc[idx]['lat']
    lon = df_top.loc[idx]['lon']
    size = 2 * df_top.loc[idx]['count']
    folium.CircleMarker(location = [lat, lon], color='rgba(171,52,40, 0.35)', radius=size).add_to(first_map)
```

```python
first_map
```

```python
for idx in df_bottom.index:
    lat = df_bottom.loc[idx]['lat']
    lon = df_bottom.loc[idx]['lon']
    folium.CircleMarker(location = [lat, lon], fill_color='rgba(17,249,199, 1)', radius=10).add_to(first_map)
```
